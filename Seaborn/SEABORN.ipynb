# import SEABORN as SNS

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
e=pd.read_csv("50ülke.csv")
e.pop("2023")
e.pop("SIRALAMA")
e.pop("SIRALAMA.2")
e
```

# distplot ( )

```python
sns.distplot(e["BÜYÜME"])
```

```python
sns.distplot(e["BÜYÜME"],
    bins=None,
    hist=True,
    kde=True,
    rug=False,
    fit=None,
    hist_kws={"histtype": "step", "linewidth": 3, "alpha": 1, "color": "r"},
    kde_kws={"color": "k", "lw": 3, "label": "KDE"},
    rug_kws=None,
    fit_kws=None,
    color='r',
    vertical=False,
    norm_hist=False,
    axlabel=None,
    label='None',
    ax=None)
```

```python
n=pd.read_csv("dünyam.csv")
n.dropna(how="any",inplace=True)
n
```

```python
sns.set(style="dark",palette="muted",font_scale=2) #for backgraound and font
sns.distplot(n["Doğum oranı"],bins=20,kde=False,color="y")
plt.tight_layout()
```

```python
sns.set(style="darkgrid",palette="muted",font_scale=1.5)
sns.distplot(n["Yıllık değişim oranı"],hist=False,rug=True,color="r")
plt.tight_layout()
```

```python
sns.set(style="white",palette="Blues",font_scale=1.5)
sns.distplot(n["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True})
plt.tight_layout()
```

```python
sns.set(style="whitegrid",palette="Blues",font_scale=1.5)
sns.distplot(n["Nüfüs"],color="m")
plt.tight_layout()
```

```python
sns.set(style="darkgrid",font_scale=1.5)

f, axes = plt.subplots(2,2,figsize=(15,10))

sns.distplot(n["Doğum oranı"],bins=20,kde=False,color="y",ax=axes[0,0])

sns.distplot(n["Yıllık değişim oranı"],hist=False,rug=True,color="r",ax=axes[0,1])

sns.distplot(n["Ortalama yaş"],hist=False,color="g",kde_kws={"shade":True},ax=axes[1,0])

sns.distplot(n["Nüfüs"],color="m",ax=axes[1,1])

plt.tight_layout()
```
